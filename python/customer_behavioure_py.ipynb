```python
# to fill the avg median value in the null cells 
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# to replace the space to _ and set the alphabet into lower case and rename the columne name purchase_amount_(usd) to purchase amount
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns ={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# to create a column age_group
```

```python
labels = ['Young Adult','Adult','Middle_aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
# create column purchase_frequency_days
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7 ,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
